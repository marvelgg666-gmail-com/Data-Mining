numpy Digunakan untuk mengelola komputasi matrix

pandas Digunakan untuk mengimpor data dari luar

```python
dataset = pd.read_csv('data.csv')
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

fungsi dari read_csv adalah untuk menghubungakan data dari luar ke dalam codingan
variabel x itu di gunakan untuk menampilakan 6 kolom variabel yang terdiri dari (ID,jenis hewan,bertelur,beranak,hidup di air,bisa terbang)
dan variabel y digunakan untuk menampilkan kolom terakhir yaitu kolom class

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer (missing_values=np.nan, strategy='mean')
imputer.fit(x[:, 1:6])
x[:, 1:6] = imputer.transform(x[:, 1:6])
```

fungsi dari SimpleImputer adalah untuk mengubah nan yang diambil dari library sklearn dan dipanggil oleh impute.
fungsi dari missing_values ialah untuk mendeteksi nan atau eror
